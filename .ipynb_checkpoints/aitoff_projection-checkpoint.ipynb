```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# all clusters
data = pd.read_csv('clusters.dat', sep=' ', header=None, skipinitialspace=True)
data.columns=['Cluster', 'RAdeg', 'DEdeg', 'Log(Age)', 'E(B-V)','pmRA', 'pmDE', 'plx','Dist', 'NBSS', 'NYSS', 'Notes']

data
```

```python
# latitude is dec 
# longitude is ra 
```

# Aitoff Projection

```python
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
cluster_ra=np.array(data['RAdeg'])
cluster_dec=np.array(data['DEdeg'])

eq = SkyCoord(cluster_ra, cluster_dec, unit=u.deg)
gal = eq.galactic
```

```python
plt.subplot(111, projection='aitoff')
plt.grid(True)
plt.scatter(gal.l.wrap_at('180d').radian, gal.b.radian, marker='*', color='pink')
plt.show()
```

# Bins

```python
data_2 = data.sort_values(by=['Dist'])
data_2
```

```python


data_2['Binned'], bins = pd.qcut(data_2['Dist'], q=[0, 0.25, 0.50, .75, 1],
                               labels=["1", "2", "3", '4'],
                               retbins=True)

data_2['gal_l'] = gal.l
data_2['gal_b'] = gal.b
data_2
```

```python
bins = np.linspace(0, 1, 4)
hist, bins = np.histogram(data['Dist'],bins=4,density=True)

bin1 = data_2[(data_2['Dist'] >= bins[0]) & (data_2['Dist'] < bins[1])].index
bin2 = data_2[(data_2['Dist'] >= bins[1]) & (data_2['Dist'] < bins[2])].index
bin3 = data_2[(data_2['Dist'] >= bins[2]) & (data_2['Dist'] < bins[3])].index
bin4 = data_2[(data_2['Dist'] >= bins[3]) & (data_2['Dist'] <= bins[4])].index


data_2['NewBin'] = data_2.index
data_2
```

```python
for x in bin1:
    data_2.loc[data_2.NewBin==x, 'NewBin']=1

for y in bin2:
    data_2.loc[data_2.NewBin==y, 'NewBin']=2
    
for z in bin3:
    data_2.loc[data_2.NewBin==z, 'NewBin']=3
    
for w in bin4:
    data_2.loc[data_2.NewBin==w, 'NewBin']=4

data_2
```
